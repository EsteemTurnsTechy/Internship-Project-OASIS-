```python
import pandas as pd

df = pd.read_csv("unzipped_data/AB_NYC_2019.csv")
df.head()
```

```python
# Overview of dataset
df.info()

# Summary statistics
df.describe(include='all')
```

```python
# Check for missing values
df.isnull().sum()

df = df.dropna()
```

```python
# Check for duplicates
print("Duplicates:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()
```

```python

df = df.drop(columns=['host_id', 'host_name' 'neighbourhood_group '], errors='ignore')
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['price'])
plt.title("Price Outliers")
plt.show()

# Remove extreme price outliers (e.g., above 99th percentile)
upper_limit = df['price'].quantile(0.99)
df = df[df['price'] < upper_limit]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set a style
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='room_type', order=df['room_type'].value_counts().index, palette="viridis")
plt.title("Distribution of Room Types")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.show()
```

```python
avg_price_neighborhood = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_price_neighborhood.values, y=avg_price_neighborhood.index, palette="magma")
plt.title("Top 10 Expensive Neighborhoods by Average Price")
plt.xlabel("Average Price")
plt.ylabel("Neighborhood")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='price', y='number_of_reviews', alpha=0.5)
plt.title("Price vs Number of Reviews")
plt.xlim(0, 1000)  # to reduce the effect of extreme outliers
plt.xlabel("Price")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

features = df[['price', 'number_of_reviews', 'availability_365', 'minimum_nights', 'latitude', 'longitude']].copy()

# Remove extreme outliers (optional)
features = features[features['price'] < 1000]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='longitude', y='latitude', hue='cluster', data=df, palette='tab10', alpha=0.6)
plt.title("Listing Clusters by Location")
plt.show()
```
